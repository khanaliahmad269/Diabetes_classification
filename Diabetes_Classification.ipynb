```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/pima-indians-diabetes.csv')
```

```python
columns_labels=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']
df.columns=columns_labels
```

###Data Visualization


```python
sns.heatmap(df.corr(), annot=True)
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
x
```

```python
# Spliting the dataset into train and test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_sc = sc.fit_transform(x_train)
X_test_sc = sc.transform(x_test)
```

```python
X_train_sc
```

```python
X_test_sc
```

###KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

# Initiating the classifier
model_knn = KNeighborsClassifier()

# Passing the data to classifier
model_knn.fit(X_train_sc, y_train)
```

```python
y_pred = model_knn.predict(X_test_sc)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics  import classification_report
knn_report = classification_report(y_test, y_pred)
```

```python
print( knn_report)
```

```python
print( DT_report)
```
